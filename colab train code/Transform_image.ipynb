```python
import os
import glob
from PIL import Image
from pathlib import Path

files = glob.glob("/content/gdrive/MyDrive/capston/dataset/train/images/*.jpg")

for f in files:
    img = Image.open(f)
    img_resize = img.resize((640,640),Image.LANCZOS)
    
    title, ext = os.path.splitext(f)
    file_name = Path(f).stem
    title="/content/gdrive/MyDrive/croll/croll2/"+file_name
    img_resize.save(title+ext)
```
